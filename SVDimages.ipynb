```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color

image = io.imread("Berkeley1.png")
A = color.rgb2gray(image) 

U, S, Vt = np.linalg.svd(A)

plt.scatter(np.arange(1, 1201), S, s=.5)
plt.title("Singular Values")
plt.xlabel("k")
plt.ylabel("Value")
plt.show()

def approximation(A, U, S, Vt, k):
    Uk = U[:, :k]
    Sk = np.diag(S[:k])
    Vk = Vt[:k, :]
    
    B_k = np.dot(np.dot(Uk, Sk), Vk)
    
    error = np.linalg.norm(A - B_k, 'fro') ** 2 / np.linalg.norm(A, 'fro') ** 2
    
    return B_k, error

ks = [30, 80, 100]
for k in ks:
    B_k, error = approximation(A, U, S, Vt, k)
    percentage_singular_values = (k / len(S)) * 100
    
    print(f"k = {k}:")
    print(f"Error (e_k) = {error:.4f}")
    print(f"Percentage of singular values used = {percentage_singular_values:.2f}%")
    
    plt.figure(figsize=(6, 6))
    plt.imshow(B_k, cmap='gray')
    plt.title(f'Rank-{k} Approximation')
    plt.axis('off')
    plt.show()
```
